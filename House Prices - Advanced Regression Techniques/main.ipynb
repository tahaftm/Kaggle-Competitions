### Encoding the categorical values

```python
categorical_col = list(train.select_dtypes(include='object').columns)
categorical_col
```

```python

len(categorical_col)
```

```python
def get_encoded_df(col):
    from sklearn.preprocessing import OneHotEncoder
    ohe = OneHotEncoder(handle_unknown='ignore',sparse_output=False)
    train_encoded_arr = ohe.fit_transform(X_train[col])
    test_encoded_arr = ohe.transform(X_test[col])
    train_encoded_df = pd.DataFrame(train_encoded_arr, columns=ohe.get_feature_names_out(), index=X_train.index)
    test_encoded_df = pd.DataFrame(test_encoded_arr, columns=ohe.get_feature_names_out(), index=X_test.index)
    return train_encoded_df,test_encoded_df
```

```python
X_train_encoded, X_test_encoded = get_encoded_df(categorical_col)
```

```python
len(ohe.get_feature_names_out())
```

```python
X_train.shape
```

```python
X_train = pd.concat([X_train, X_train_encoded], axis = 1)
```

```python
X_train.drop(categorical_col,axis = 1,inplace=True)
```

```python
X_test = pd.concat([X_test, X_test_encoded], axis = 1)
X_test.drop(categorical_col,axis = 1,inplace=True)
 
```
